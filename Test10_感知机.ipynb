```python
# 感知机原理：任何像素的重要性都以复杂的方式取决于该像素的上下文（周围像素的值）。
'''隐藏层：在仿射变换之后对每个隐藏单元应用非线性的激活函数（activation function）。 激活函数的输出被称为活性值（activations）。
   一般来说，有了激活函数，就不可能再将我们的多层感知机退化成线性模型'''
```

```python
# 激活函数是深度学习的基础
%matplotlib inline
import torch
from d2l import torch as d2l
```

```python
# 1.ReLU函数
x = torch.arange(-4.0, 4.0, 0.1, requires_grad=True)
y = torch.relu(x)
# retain_graph=True 表示在计算完一次反向传播后保留计算图。这在需要多次反向传播时是必要的
# 使用y.sum()的方法更加简洁，因为它可以直接计算y中所有元素的和，并将其作为一个标量来处理，不需要手动指定梯度权重。
y.backward(torch.ones_like(x), retain_graph=True)

d2l.plot(X= x.detach(), Y= [y.detach(), x.grad], xlabel= 'x', figsize=(5, 2.5), legend=['relu(x)', 'grad of relu'])
```

```python
# 2.Sigmoid函数
y = torch.sigmoid(x)
# 清除以前的梯度
x.grad.zero_()
y.backward(torch.ones_like(x),retain_graph=True)

d2l.plot(x.detach(), [y.detach(), x.grad], 'x', figsize=(5, 2.5), legend= ['sigmoid(x)', 'grad of sigmoid'])
```

```python
# 3.tanh函数
y = torch.tanh(x)
# 清除以前的梯度
x.grad.zero_()
y.backward(torch.ones_like(x),retain_graph=True)

d2l.plot(x.detach(), [y.detach(), x.grad], 'x', figsize=(5, 2.5), legend= ['tanh(x)', 'grad of tanh'])
```

```python
import torch
from torch import nn
from d2l import torch as d2l

batch_size = 10
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)

# for X, _ in test_iter: # test_iter未经过混乱，数据稳定
#     break;
# # 预览Resize+ToTensor后的图像。
# d2l.show_images(X.reshape(batch_size, 28,28), 2, 5)
```

```python
# 初始化模型参数(hidden:隐藏层参数2^a)
num_inputs, num_outputs, num_hiddens = 784, 10, 256
# 1.H=F(XW1+b1),O= HW2+b2; F是激活函数，X是输入n*num_inputs, O是输出n*num_outputs.
# 2.randn:元素符合标准正态分布且在(0,1); *0.01使得权重初始值的范围变得较小，使模型更容易收敛，提高训练效果。
W1 = nn.Parameter(torch.randn(num_inputs, num_hiddens, requires_grad=True) * 0.01)
b1 = nn.Parameter(torch.zeros(num_hiddens, requires_grad=True))

W2 = nn.Parameter(torch.randn(num_hiddens, num_outputs, requires_grad=True) * 0.01)
b2 = nn.Parameter(torch.zeros(num_outputs, requires_grad=True))

params = [W1, b1, W2, b2]

# 自定义relu：逐元素比较张量 X 和 a 的值，并返回对应位置的最大值
def relu(X):
    a = torch.zeros_like(X)
    return torch.max(X, a)

# 定义网络结构
def net(X):
    X = X.reshape((-1, num_inputs))
    H = relu(X@W1 + b1)  # 在 PyTorch 中，@ 符号表示矩阵乘法操作
    return (H@W2 + b2)

# 交叉熵损失函数
loss = nn.CrossEntropyLoss(reduction='none')
```

```python
num_epochs, lr = 10, 0.1
updater = torch.optim.SGD(params, lr=lr)
d2l.train_ch3(net, train_iter, test_iter, loss, num_epochs, updater)
```

```python
d2l.predict_ch3(net, test_iter)
```

```python
import torch
from torch import nn
from d2l import torch as d2l

'''模型包含：展平层，全连接，激活函数，全连接'''
net = nn.Sequential(nn.Flatten(), # n*input
                    nn.Linear(784, 256), # input*hidden
                    nn.ReLU(), # 破坏线性关系
                    nn.Linear(256, 10)) # hidden*output
'''初始化每一个全连接层的参数'''
def init_weights(m):
    if type(m) == nn.Linear:
        # 确保了只有 nn.Linear 层的权重会被初始化。m.weight:W; m.bias: b
        nn.init.normal_(m.weight, std=0.01)
        if m.bias is not None:
            # 通常也会初始化偏置
            nn.init.zeros_(m.bias)

#  apply 会遍历 net 中的所有子模块，并对每个子模块调用 init_weights 函数
net.apply(init_weights);

'''训练模型'''
batch_size, lr, num_epochs = 256, 0.1, 10
loss = nn.CrossEntropyLoss(reduction='none')
updater = torch.optim.SGD(net.parameters(), lr=lr)

train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
d2l.train_ch3(net, train_iter, test_iter, loss, num_epochs, updater)
```

```python
d2l.predict_ch3(net, test_iter)
```
