```python
'''[权重衰减是一种特定的 L2 正则化方法]，通过在优化过程中直接对权重进行衰减来实现。
   它通过在每次参数更新时减少一定比例的权重值来防止[过拟合]。'''
```

```python
%matplotlib inline
import torch
from torch import nn
from d2l import torch as d2l
```

```python
n_train, n_test, num_inputs, batch_size = 20, 100, 200, 5
true_w, true_b = torch.ones((num_inputs, 1)) * 0.01, 0.05 # true_w:200*1
# train_data= (X,y)--(20*200)=(n_train*len(true_w)),true_b---(20*1); y=matmul(X,w)+b+ε.
train_data = d2l.synthetic_data(true_w, true_b, n_train)
train_iter = d2l.load_array(train_data, batch_size)
test_data = d2l.synthetic_data(true_w, true_b, n_test)
test_iter = d2l.load_array(test_data, batch_size, is_train=False)
```

```python
# 随机初始化模型参数(需要求的结果)
def init_params():
    w = torch.normal(0, 1, size=(num_inputs, 1), requires_grad=True)
    b = torch.zeros(1, requires_grad=True)
    return [w, b]

# L2范数^2作为惩罚,只差一个参数 λ。
def l2_penalty(w):
    return torch.sum(w.pow(2)) / 2
```

```python
def train(lambd):
    w, b = init_params()
    # 3.2 linreg：Xw+b，squared_loss：均方损失
    net, loss = lambda X: d2l.linreg(X, w, b), d2l.squared_loss
    num_epochs, lr = 100, 0.003
    animator = d2l.Animator(xlabel='epochs', ylabel='loss', yscale='log',
                            xlim=[5, num_epochs], legend=['train', 'test'])
    for epoch in range(num_epochs):
        for X, y in train_iter:
            # 增加了L2范数惩罚项，
            # 广播机制使l2_penalty(w)成为一个(batch_size,1)的向量
            l = loss(net(X), y) + lambd * l2_penalty(w)
            l.sum().backward()
            d2l.sgd([w, b], lr, batch_size)
        if (epoch + 1) % 5 == 0:
            animator.add(epoch + 1, (d2l.evaluate_loss(net, train_iter, loss),
                                     d2l.evaluate_loss(net, test_iter, loss)))
    # norm:L2范数返回tensor标量张量，item：将张量中的单个值转换为一个 Python 标量
    print('w的L2范数是：', torch.norm(w).item())
```

```python
# 不使用权重衰减，出现了严重的过拟合。
train(lambd=0)
```

```python
# 我们使用权重衰减来运行代码。 注意，在这里训练误差增大，但测试误差减小。 这正是我们期望从正则化中得到的效果。
train(lambd=3)
```

```python
'''权重衰减简洁实现'''
```

```python
def train_concise(wd):
    net = nn.Sequential(nn.Linear(num_inputs, 1))
    for param in net.parameters():
        param.data.normal_()
    loss = nn.MSELoss(reduction='none')
    num_epochs, lr = 100, 0.003
    # wd代表权重衰减
    updater = torch.optim.SGD([
        {"params":net[0].weight,'weight_decay': wd},
        {"params":net[0].bias}],
        lr=lr)
    animator = d2l.Animator(xlabel='epochs', ylabel='loss', yscale='log',
                            xlim=[5, num_epochs], legend=['train', 'test'])
    for epoch in range(num_epochs):
        for X, y in train_iter:
            updater.zero_grad() # 清零梯度
            l = loss(net(X), y)
            l.mean().backward() # 反向传播计算梯度
            updater.step() # 更新模型参数
        if (epoch + 1) % 5 == 0:
            animator.add(epoch + 1,
                         (d2l.evaluate_loss(net, train_iter, loss),
                          d2l.evaluate_loss(net, test_iter, loss)))
    print('w的L2范数：', net[0].weight.norm().item())
```

```python
# 不使用权重衰减，出现了严重的过拟合。
train_concise(0)
```

```python
# 使用权重衰减
train_concise(3)
```

```python
'''暂退法实现<mxnet与numpy(d2l依赖)存在版本冲突，此处放弃运行>'''
'''简单性的另一个角度是平滑性，即函数不应该对其输入的微小变化敏感。当我们对图像进行
   分类时，我们预计向像素添加一些随机噪声应该是基本无影响的。在计算后续层之前向网络
   的每一层注入噪声就称为[暂退法]。因为当训练一个有多层的深层网络时，注入噪声只会在
   输入-输出映射上增强平滑性。'''
```

```python
from mxnet import autograd, gluon, init, np, npx
from mxnet.gluon import nn
from d2l import mxnet as d2l

npx.set_np()

# 以dropout为阈值，小于阈值的随机值，丢弃X对应的元素。
def dropout_layer(X, dropout):
    assert 0 <= dropout <= 1
    if dropout == 1:
        return np.zeros_like(X)
    if dropout == 0:
        return X
    # uniform:生成随机数0~1间。mask:0,1的np_array
    mask = np.random.uniform(0, 1, X.shape) > dropout
    '''假设 X 中的每个元素都是独立随机变量，并且在应用 dropout 后，每个元素被保留的概率为 1 - dropout。
       这意味着保留下来的值的期望是原始值的 1 - dropout 倍。所以，为了补偿这一点，我们需要将保留下来的值
       除以 1 - dropout，以保持期望值不变。'''
    return mask.astype(np.float32) * X / (1.0 - dropout)
```

```python
# 测试函数
X = np.arange(16).reshape(2, 8)
print(dropout_layer(X, 0))
print(dropout_layer(X, 0.5))
print(dropout_layer(X, 1))
```

```python
# 设计多层网络
num_inputs, num_outputs, num_hiddens1, num_hiddens2 = 784, 10, 256, 256

dropout1, dropout2 = 0.2, 0.5

class Net(nn.Module):
    def __init__(self, num_inputs, num_outputs, num_hiddens1, num_hiddens2, is_training = True):
        super(Net, self).__init__()
        self.num_inputs = num_inputs
        self.training = is_training
        # 全连接层：XW+b。X:batch*in, W:in*hidden, b:1*hidden
        self.lin1 = nn.Linear(num_inputs, num_hiddens1)
        self.lin2 = nn.Linear(num_hiddens1, num_hiddens2)
        self.lin3 = nn.Linear(num_hiddens2, num_outputs)
        self.relu = nn.ReLU()

    # 全连接-隐藏relu-暂退法dropout-全连接-输出
    def forward(self, X):
        # relu非线性变换代表了隐藏层：隐藏层的目的是在输入数据和最终输出之间引入非线性变换
        H1 = self.relu(self.lin1(X.reshape((-1, self.num_inputs))))
        # 只有在训练模型时才添加dropout层
        if self.training == True:
            H1 = dropout_layer(H1, dropout1)
        H2 = self.relu(self.lin2(H1))
        if self.training == True:
            H2 = dropout_layer(H2, dropout2)
        out = self.lin3(H2)
        return out


net = Net(num_inputs, num_outputs, num_hiddens1, num_hiddens2)
```

```python
num_epochs, lr, batch_size = 10, 0.5, 256
loss = nn.CrossEntropyLoss(reduction='none')
# for X,y in iter:
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
updater = torch.optim.SGD(net.parameters(), lr=lr)
d2l.train_ch3(net, train_iter, test_iter, loss, num_epochs, updater)
```
