```python
import torch
''''''
def synthetic_data(w, b, num_examples):  #@save
    """y=Xw+b+"""
    X = torch.normal(0, 1, (num_examples, len(w)))
    # mvXw
    y = torch.matmul(X, w) + b
    y += torch.normal(0, 0.01, y.shape)
    return X, y.reshape((-1, 1))

true_w = torch.tensor([2, -3.4])
true_b = 4.2
# features:Xlabelsy
features, labels = synthetic_data(true_w, true_b, 1000)

# 
%matplotlib inline
import random
import torch
from d2l import torch as d2l
d2l.set_figsize()
d2l.plt.xlabel('feature(X[0],X[1])') # 
d2l.plt.ylabel('label(y=Xw+b+)') # 
d2l.plt.scatter(features[:, 0].detach().numpy(), labels.detach().numpy(), 1); # labels[i]= features[i][0]*w[0]+b+
d2l.plt.scatter(features[:, 1].detach().numpy(), labels.detach().numpy(), 1); # labels[i]= features[i][1]*w[1]+b+
```

```python
print(list(range(4))) # arangerange
print(torch.arange(4))

def data_iter(batch_size, features, labels): # featureslabels
    '''sizebatch'''
    num_examples = len(features)
#     indices = list(range(num_examples))
    indices = torch.arange(num_examples) # 0~num-1
    random.shuffle(indices) # 
    for i in range(0, num_examples, batch_size): # for i in [0batch_size,2*batch_size...]
        batch_indices = indices[i: min(i + batch_size, num_examples)] # indicesi~i+batch_size
        # (batch_size, num_features)(batch_size,) yield  yield 
        yield features[batch_indices], labels[batch_indices]
```

```python
# ()
batch_size = 10
print(features.shape, '\n', labels.shape)

for X, y in data_iter(batch_size, features, labels):
    print(X, '\n', y) # 1000/10=100
    break
```

```python
print(features.shape, labels.shape, 'batch_size:', batch_size, '\nw^T:', true_w.T, 'b:', true_b)

# 1.
w = torch.normal(0, 0.01, size=(2,1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
print('init w.T,b:', w.T, b)

# 2.
def linreg(X, w, b):  #@save
    """y= Xw+b"""
    return torch.matmul(X, w) + b

# 3.(y_hat y )
def squared_loss(y_hat, y):  #@save
    """"""
    return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2 # y.reshape(y_hat.shape)

# 4.
def sgd(params, lr, batch_size):  #@save
    """mini-batch"""
    with torch.no_grad(): # 
        for param in params: # [w,b]
#             if len(param)>1:
#                 print('grad:', param.grad.T)
            # 3.1.1.2 :1/(2n)(y-yi)^2=1/(2n)(Xw+b-yi)^2, .
            param -= lr * param.grad / batch_size
            param.grad.zero_()
            
# 5.
lr = 0.03 # 
num_epochs = 3 # 
net = linreg

for epoch in range(num_epochs):
    for X, y in data_iter(batch_size, features, labels):
        '''X:(batch_size,feature_num); w(feature_num,); y(batch_size,);'''
        l = squared_loss(net(X, w, b), y)  # Xy
        # l(batch_size,1)l
        # [w,b]
        l.sum().backward()
        sgd([w, b], lr, batch_size)  # 

    # 
    with torch.no_grad():
#         print(torch.cat((net(features, w, b), labels), dim=1)) # labels
        train_l = squared_loss(net(features, w, b), labels) # 
        # :f 
        print(f'epoch {epoch + 1},loss {float(train_l.mean()):f}, w is {w.T}, b is {b}')
```
