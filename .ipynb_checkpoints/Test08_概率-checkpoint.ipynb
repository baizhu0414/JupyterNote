```python
%matplotlib inline
import torch
from torch.distributions import multinomial
from d2l import torch as d2l
```

```python
fair_probs = torch.ones([6]) / 6
counts = multinomial.Multinomial(1000, fair_probs).sample()
counts/1000, 1/6
```

```python
counts = multinomial.Multinomial(10, fair_probs).sample((500,))
cum_counts = counts.cumsum(dim=0) # 累积和，dim=0从上往下加
print(counts, '\n', cum_counts)
estimates = cum_counts / cum_counts.sum(dim=1, keepdims=True)
print(cum_counts.sum(dim=1, keepdims=True)[:12,], '......')

d2l.set_figsize((6, 4.5))
for i in range(6): # 0~5
    d2l.plt.plot(estimates[:, i].numpy(),
                 label=("P(die=" + str(i + 1) + ")"))

d2l.plt.axhline(y=0.167, color='black', linestyle='dashed')
d2l.plt.gca().set_xlabel('Groups of experiments')
d2l.plt.gca().set_ylabel('Estimated probability')
d2l.plt.legend();
```

```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
import torch
import math

# 正态分布的概率密度函数，不同于torch.normal。
def normalY(mu, sigma, x):
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 / sigma**2 * (x - mu)**2)
'''绘制正态分布函数图像'''
# x= np.arange(-2, 2, 0.1)
x = np.linspace(-2, 2, 100)
# x = torch.normal(0, 1, size= (100,))
y = normalY(0, 1, x)
z = normalY(1, 1, x)
h = normalY(0, 0.5, x)

fig, ax= plt.subplots(1)
# ax.scatter(x, 4*torch.rand(len(x))-2, label='normal x distribution', alpha=0.5)
# ax.axvline(x=1, linestyle='--', label='x = 1') # 绘制x=1线条
# ax.axvline(x=-1, linestyle='--', label='x = -1')
# 此处不能重复调用x1,y1，而应该用x和np.exp代替，或者用y.detach()定住，否则报错。
ax.plot(x, y, label='x~N(0,1)', linestyle='-')
ax.plot(x, z, label='x~N(1,1)', linestyle='--')
ax.plot(x, h, label='x~N(0,0.5)', linestyle='-.')
ax.legend(shadow=True)
plt.title('normal Y')

plt.show()
```
